DistilBERT is a streamlined version of the BERT model, designed for faster training and inferencing on off-cloud servers, making it a suitable choice for applications with constraints on computational resources (Wei et al., 2022). As a distilled version of BERT, DistilBERT maintains a significant proportion of the original model's capabilities, including 97% of BERT's language understanding proficiency, while offering the advantages of being 60% faster and 40% smaller (Sanh et al., 2019).

The model has demonstrated its efficacy across various applications, such as suicidal intention detection in social media posts (Ananthakrishnan et al., 2022) and opinion holder detection (Al-Mahmud & Shimada, 2022). This versatility is attributable to its robust foundational architecture derived from BERT, coupled with enhancements in efficiency and speed.

```python
data = pd.read_json("../../data/dataset_english.json")
data
```

```python
def convert_privacy_mask(privacy_mask):
    """
    Converts the privacy mask dictionary in a format that is easier to process. 
    It removes brackets and splits by underscores to extract the primary entity name.

    :param privacy_mask: Dictionary where keys are entity names with additional characters.
    :return: A new dictionary with formatted entity names.
    """
    new_dict = {}
    for key, value in privacy_mask.items():
        # Remove brackets and split by underscore to get the primary entity name
        new_key = key.strip('[]').split('_')[0]
        new_dict[new_key] = value
    return new_dict

# Apply the conversion function to the 'privacy_mask' column to create 'data_results'
data['data_results'] = data['privacy_mask'].apply(convert_privacy_mask)

data[['privacy_mask', 'data_results']].head()
```

```python
data_small = data
```

```python
# An empty list to store the simplified results
model_simplified_results = []

# Through the rows of the df
for index, row in data_small.iterrows():
    text = row['unmasked_text']
    # Apply the model to the text
    result = nlp(text)

    # Simplify the model output and extract entity and word
    simplified_dict = {}
    for item in result:
        # Remove the "B-" or "I-" or "-O" prefix from the entity label
        entity = item['entity'][2:]
        word = item['word']
        # Accumulate words for each entity, separating them with space
        simplified_dict[entity] = simplified_dict.get(entity, "") + " " + word

    # Strip extra spaces from the accumulated words for each entity
    simplified_dict = {k: v.strip() for k, v in simplified_dict.items()}

    # Append the simplified dictionary to the list
    model_simplified_results.append(simplified_dict)

# Add new column with results of the model
data_small['model_results'] = model_simplified_results
```

```python
# Extract all unique entities from the 'data_results' column
unique_entities = set()
for _, row in data_small.iterrows():
    # Update the set with entity keys from each row's 'data_results'
    unique_entities.update(row['data_results'].keys())

# Convert the set of unique entities into a list
entity_types = list(unique_entities)

# entity_types now contains a list of all unique entities found in 'data_results' across the dataset
```

```python
def evaluate_entry(true_entities, predicted_entities, entity_types):
    """
    Evaluates a single entry for true and predicted entities, calculating the number of true positives, false positives, and false negatives.

    :param true_entities: Dictionary of actual entities and their values.
    :param predicted_entities: Dictionary of entities predicted by the model and their values.
    :param entity_types: List of all entity types to be evaluated.
    :return: Three Counters for true positives, false positives, and false negatives.
    """
    tp, fp, fn = Counter(), Counter(), Counter()

    for entity in entity_types:
        true_value = true_entities.get(entity, '').lower()  # Fetch and lowercase the actual entity value
        predicted_value = predicted_entities.get(entity, '').lower()  # Fetch and lowercase the predicted entity value

        if true_value and true_value == predicted_value:
            tp[entity] += 1  # True positive: Entity value is correctly predicted
        elif predicted_value:
            fp[entity] += 1  # False positive: Model predicted an entity incorrectly
        if true_value and true_value != predicted_value:
            fn[entity] += 1  # False negative: Model missed an actual entity

    return tp, fp, fn

# Counters for overall true positives, false positives, and false negatives
total_tp, total_fp, total_fn = Counter(), Counter(), Counter()

# Iterate through each entry in the dataset
for index, row in data_small.iterrows():
    tp, fp, fn = evaluate_entry(row['data_results'], row['model_results'], entity_types)
    # Results for overall metrics
    total_tp += tp
    total_fp += fp
    total_fn += fn
total_tp, total_fp, total_fn
```

```python
# Store metrics for each entity
entity_metrics = {}
for entity in entity_types:
    precision = total_tp[entity] / (total_tp[entity] + total_fp[entity]) if total_tp[entity] + total_fp[entity] > 0 else 0
    recall = total_tp[entity] / (total_tp[entity] + total_fn[entity]) if total_tp[entity] + total_fn[entity] > 0 else 0
    f1 = 2 * precision * recall / (precision + recall) if precision + recall > 0 else 0
    entity_metrics[entity] = {'precision': precision, 'recall': recall, 'f1_score': f1}

entity_metrics
```

```python
# Overall metrics
overall_precision = sum(total_tp.values()) / (sum(total_tp.values()) + sum(total_fp.values())) if sum(total_tp.values()) + sum(total_fp.values()) > 0 else 0
overall_recall = sum(total_tp.values()) / (sum(total_tp.values()) + sum(total_fn.values())) if sum(total_tp.values()) + sum(total_fn.values()) > 0 else 0
overall_f1 = 2 * overall_precision * overall_recall / (overall_precision + overall_recall) if overall_precision + overall_recall > 0 else 0

# Metrics for each entity
for entity, metrics in entity_metrics.items():
    print(f"Entity: {entity}, Precision: {metrics['precision']:.2f}, Recall: {metrics['recall']:.2f}, F1-Score: {metrics['f1_score']:.2f}")

# Overall metrics
print(f"Overall Precision: {overall_precision:.2f}, Overall Recall: {overall_recall:.2f}, Overall F1-Score: {overall_f1:.2f}")
```
